```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
NFLX = wb.DataReader('NFLX', data_source='yahoo', start='2005-1-1')
```

```python
NFLX.tail()
```

```python
NFLX.head()
```

```python
NFLX['log_return'] = np.log(NFLX['Adj Close']/ NFLX['Adj Close'].shift(1))
print(NFLX['log_return'])
```

```python
NFLX['log_return'].plot(figsize=(8,5))
plt.show()
```

```python
log_return_d = NFLX['log_return'].mean()
log_return_d
```

```python
log_return_a = NFLX['log_return'].mean() * 250
log_return_a
```

```python
print (str(round(log_return_a, 4)* 100) + ' %')
```
