```python
import numpy as np
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
FRO = wb.DataReader('FRO', data_source='yahoo', start='1995-1-1')
```

```python
FRO.head
```

```python
FRO.tail()
```

```python
FRO['simple_return'] = (FRO['Adj Close']/FRO['Adj Close'].shift(1)) - 1
print (FRO['simple_return'])
```

```python
FRO['simple_return'].plot(figsize=(8, 5))
plt.show()
```

```python
avg_returns_d = FRO['simple_return'].mean()
avg_returns_d
```

```python
avg_returns_a = FRO['simple_return'].mean()* 250
avg_returns_a
```

```python
print (str(round(avg_returns_a, 5)*100) + ' %')
```
